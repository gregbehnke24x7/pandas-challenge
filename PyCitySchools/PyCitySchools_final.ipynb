```python
import pandas as pd
# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
# Combine the data into a single dataset.  
comb_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
dist_tot_schools = len(pd.unique(comb_df["School ID"]))
dist_tot_students = len(pd.unique(comb_df["Student ID"]))
schools_df = pd.DataFrame(school_data)
dist_tot_budget = schools_df["budget"].sum()
dist_avg_math = comb_df["math_score"].mean()
dist_avg_read = comb_df["reading_score"].mean()
dist_tot_math_pass =  comb_df.loc[comb_df["math_score"] >= 70]["Student ID"].count()
dist_tot_read_pass =  comb_df.loc[comb_df["reading_score"] >= 70]["Student ID"].count()
dist_tot_compl_pass = comb_df.loc[(comb_df["math_score"] >= 70) & (comb_df["reading_score"] >= 70)]["Student ID"].count()
dist_math_pass_pct = dist_tot_math_pass / dist_tot_students * 100
dist_read_pass_pct = dist_tot_read_pass / dist_tot_students * 100
dist_compl_pass_pct = dist_tot_compl_pass / dist_tot_students * 100
dist_df = [{"School Count": dist_tot_schools,
            "Student Count": dist_tot_students,
            "District Budget": dist_tot_budget,
            "Average Math Score": dist_avg_math,
            "Average Reading Score": dist_avg_read,
            "Students Passing Math": dist_tot_math_pass,
            "Students Passing Reading": dist_tot_read_pass,
            "Students Passing Both": dist_tot_compl_pass,
            "Dist. Math Pass %": dist_math_pass_pct,
            "Dist. Read Pass %": dist_read_pass_pct,
            "Dist. Overall Pass %": dist_compl_pass_pct}]
        
dist_df = pd.DataFrame(dist_df)
dist_df["District Budget"] = dist_df["District Budget"].map("${:,.2f}".format)
dist_df["Average Math Score"] = dist_df["Average Math Score"].map("{:,.2f}".format)
dist_df["Average Reading Score"] = dist_df["Average Reading Score"].map("{:,.2f}".format)
dist_df["Dist. Math Pass %"] = dist_df["Dist. Math Pass %"].map("{:,.2f}%".format)
dist_df["Dist. Read Pass %"] = dist_df["Dist. Read Pass %"].map("{:,.2f}%".format)
dist_df["Dist. Overall Pass %"] = dist_df["Dist. Overall Pass %"].map("{:,.2f}%".format)
dist_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
schools_df.rename(columns = {"type":"Type", "size":"Enrollment", "budget": "Budget"}, inplace=True)
schools_df.set_index("school_name", inplace=True)
schools_df["Per Student Budget"] = schools_df["Budget"] / schools_df["Enrollment"]
schools_df["Average Math Score"] = comb_df.groupby("school_name")["math_score"].mean()
schools_df["Average Reading Score"] = comb_df.groupby("school_name")["reading_score"].mean()
schools_df["Math Pass %"] = comb_df[comb_df["math_score"] >= 70].groupby("school_name")["math_score"].count() / schools_df["Enrollment"] * 100
schools_df["Reading Pass %"] = comb_df[comb_df["reading_score"] >= 70].groupby("school_name")["reading_score"].count() / schools_df["Enrollment"] * 100
schools_df["Overall Pass %"] = \
comb_df.loc[(comb_df["math_score"] >= 70) & (comb_df["reading_score"] >= 70)].groupby("school_name")["school_name"].count() / schools_df["Enrollment"] * 100
schools_df.sort_values("school_name", axis=0, ascending=True, inplace=True)
schools_fmt_df = schools_df.copy()
schools_fmt_df["Enrollment"] = schools_fmt_df["Enrollment"].map("{:,.0f}".format)
schools_fmt_df["Budget"] = schools_fmt_df["Budget"].map("${:,.2f}".format)
schools_fmt_df["Per Student Budget"] = schools_fmt_df["Per Student Budget"].map("${:,.2f}".format)
schools_fmt_df["Average Math Score"] = schools_fmt_df["Average Math Score"].map("{:,.2f}".format)
schools_fmt_df["Average Reading Score"] = schools_fmt_df["Average Reading Score"].map("{:,.2f}".format)
schools_fmt_df["Math Pass %"] = schools_fmt_df["Math Pass %"].map("{:,.2f}%".format)
schools_fmt_df["Reading Pass %"] = schools_fmt_df["Reading Pass %"].map("{:,.2f}%".format)
schools_fmt_df["Overall Pass %"] = schools_fmt_df["Overall Pass %"].map("{:,.2f}%".format)
schools_fmt_df.head()
```

## Math Scores by Grade

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
schools_df.reset_index(inplace=True)
bins = [0, 585, 630, 645, 680]
labels = ["<$\$$585", "$\$$585-$\$$630", "$\$$630-$\$$645", "$\$$645-$\$$680"]

schools_df["Budget Group"] = pd.cut(schools_df["Per Student Budget"], bins, labels=labels)
schools_data = [schools_df["school_name"], schools_df["Average Math Score"], schools_df["Average Reading Score"],\
              schools_df["Math Pass %"], schools_df["Reading Pass %"], schools_df["Overall Pass %"], schools_df["Budget Group"]]
schools_headers = ["School Name", "Average Math Score", "Average Reading Score", "Math Pass %", "Reading Pass %", "Overall Pass %", "Budget Group"]
schools_pb = pd.concat(schools_data, axis=1, keys=schools_headers) 
schools_pb = pd.DataFrame(schools_pb)
out_pb_df = schools_pb.groupby("Budget Group").mean()
out_pb_df = pd.DataFrame(out_pb_df)
out_pb_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
bins = [0, 1000, 2000, 5000]
labels = ["Small(<1000)", "Medium(1000-2000)", "Large(2000-5000)"]
schools_df["Size Group"] = pd.cut(schools_df["Enrollment"], bins, labels=labels)
schools_data = [schools_df["school_name"], schools_df["Average Math Score"], schools_df["Average Reading Score"],\
                schools_df["Math Pass %"], schools_df["Reading Pass %"], schools_df["Overall Pass %"], schools_df["Size Group"]]
schools_headers = ["School Name", "Average Math Score", "Average Reading Score", "Math Pass %", "Reading Pass %", "Overall Pass %", "Size Group"]
schools_sz = pd.concat(schools_data, axis=1, keys=schools_headers) 
schools_sz = pd.DataFrame(schools_sz)
out_sz_df = schools_sz.groupby("Size Group").mean()
out_sz_df = pd.DataFrame(out_sz_df)
out_sz_df
```
